```python

import numpy as np

import matplotlib.pyplot as plt 
from matplotlib import animation 
from IPython.display import HTML

fig, ax= plt.subplots() 
ax.set_xlim(0, 10)

ax.set_ylim(0, 10)

#Initial Line data 
x= np.linspace(0, 10, 100)

y= np.zeros_like(x)

line, = ax.plot(x, y, lw-2, color='red')

def animate(frame):
    
    y = np.sin(x +frame 0.1) + 5
    
    line.set_data(x, y)

    return (line,)

anim = animation.FuncAnimation(

    animate, 
    frames-100, 
    interval-50,
    blit-True

HTML(anim.to_jshtml())
```
