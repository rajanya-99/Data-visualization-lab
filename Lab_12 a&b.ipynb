```python
import sys
!{sys.executable} -m pip install pygame
import pygame
import sys

pygame.init()

# Screen size
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Animated Circle with Keyboard Control")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BLUE = (100, 150, 255)

# Circle properties
x, y = WIDTH // 2, HEIGHT // 2
radius = 20
speed = 5

clock = pygame.time.Clock()

while True:
    screen.fill(WHITE)

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Key press handling
    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT]:
        x -= speed
    if keys[pygame.K_RIGHT]:
        x += speed
    if keys[pygame.K_UP]:
        y -= speed
    if keys[pygame.K_DOWN]:
        y += speed

    # Draw axes
    pygame.draw.line(screen, BLACK, (0, HEIGHT // 2), (WIDTH, HEIGHT // 2), 2)  # X-axis
    pygame.draw.line(screen, BLACK, (WIDTH // 2, 0), (WIDTH // 2, HEIGHT), 2)  # Y-axis

    # Labels
    font = pygame.font.SysFont(None, 24)
    x_label = font.render("X-axis", True, BLACK)
    y_label = font.render("Y-axis", True, BLACK)

    screen.blit(x_label, (WIDTH - 80, HEIGHT // 2 + 10))
    screen.blit(y_label, (WIDTH // 2 + 10, 10))

    # Draw circle
    pygame.draw.circle(screen, BLUE, (x, y), radius)

    pygame.display.flip()
    clock.tick(60)
```
