```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset (use your correct file path)
df = pd.read_csv("/Users/moumitasaha/Downloads/data.csv")

# Drop unnecessary columns
df = df.drop(columns=["id", "diagnosis"])

# Take only 30 rows and 30 columns
df = df.iloc[:30, :30]

# Select two correlated features
x = df['radius_mean']
y = df['perimeter_mean']

# --- Create clusters ---
x1 = x[:15]
y1 = y[:15]

x2 = x[15:30]
y2 = y[15:30]

# --- Add outliers ---
x_out = [x.mean() + 10, x.mean() - 5]
y_out = [y.mean() + 50, y.mean() - 50]

# --- Scatter plot ---
plt.scatter(x1, y1, label="Cluster 1")
plt.scatter(x2, y2, label="Cluster 2")
plt.scatter(x_out, y_out, color='red', label="Outliers")

plt.title("Scatter Plot with Clusters, Correlation, and Outliers")
plt.xlabel("Radius Mean")
plt.ylabel("Perimeter Mean")

plt.legend()
plt.show()
```
