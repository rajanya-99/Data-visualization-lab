```python
import matplotlib.pyplot as plt
import numpy as np

# Select two features (columns)
x = df.iloc[:, 0]   
y = df.iloc[:, 1]   

# --- Create clusters ---
x1 = x[:15]
y1 = y[:15]

x2 = x[15:30]
y2 = y[15:30]

# --- Add outliers ---
x_outliers = [x.mean() + 3, x.mean() - 3]
y_outliers = [y.mean() + 5, y.mean() - 5]

# --- Plot ---
plt.scatter(x1, y1, label="Cluster 1")
plt.scatter(x2, y2, label="Cluster 2")
plt.scatter(x_outliers, y_outliers, color='red', label="Outliers")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot with Clusters, Correlation, and Outliers")

plt.legend()
plt.show()
```
