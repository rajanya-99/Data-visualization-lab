```python
import pygame
import sys
import math

pygame.init()

# Screen setup
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Circle Movement with Speed & FPS")

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
LIGHT_GRAY = (200, 200, 200)
BLUE = (50, 100, 255)

clock = pygame.time.Clock()
font = pygame.font.SysFont(None, 24)

# Circle properties
x, y = WIDTH // 2, HEIGHT // 2
vx, vy = 0, 0   # velocity
speed_value = 0

ACCEL = 0.5
FRICTION = 0.9

# Draw grid
def draw_grid():
    spacing = 30
    for i in range(0, WIDTH, spacing):
        pygame.draw.line(screen, LIGHT_GRAY, (i, 0), (i, HEIGHT))
    for j in range(0, HEIGHT, spacing):
        pygame.draw.line(screen, LIGHT_GRAY, (0, j), (WIDTH, j))

# Draw axes
def draw_axes():
    pygame.draw.line(screen, BLACK, (0, HEIGHT//2), (WIDTH, HEIGHT//2), 2)
    pygame.draw.line(screen, BLACK, (WIDTH//2, 0), (WIDTH//2, HEIGHT), 2)

    screen.blit(font.render("X-axis", True, BLACK), (WIDTH - 80, HEIGHT//2 + 10))
    screen.blit(font.render("Y-axis", True, BLACK), (WIDTH//2 + 10, 10))


while True:
    screen.fill(WHITE)

    draw_grid()
    draw_axes()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Key handling (acceleration based)
    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT]:
        vx -= ACCEL
    if keys[pygame.K_RIGHT]:
        vx += ACCEL
    if keys[pygame.K_UP]:
        vy -= ACCEL
    if keys[pygame.K_DOWN]:
        vy += ACCEL

    # Apply friction (smooth stop)
    vx *= FRICTION
    vy *= FRICTION

    # Update position
    x += vx
    y += vy

    # Speed calculation
    speed_value = math.sqrt(vx**2 + vy**2)

    # Draw circle
    pygame.draw.circle(screen, BLUE, (int(x), int(y)), 10)

    # FPS
    fps = clock.get_fps()

    # Display info
    speed_text = font.render(f"Speed: {speed_value:.2f}", True, BLACK)
    fps_text = font.render(f"FPS: {int(fps)}", True, BLACK)

    screen.blit(speed_text, (10, 10))
    screen.blit(fps_text, (10, 40))

    pygame.display.flip()

    clock.tick(60)
```
